```python
import pandas as pd
import json
import os
import cv2
from collections import defaultdict
from ultralytics import YOLO
```

```python
with open("FLIR_ADAS_v2/images_thermal_train/coco.json", 'r') as file:
    cocoData = json.load(file)

```

```python
images = cocoData['images']
annotations = cocoData['annotations']
cats = cocoData['categories']
catMap = {cat['id']: cat['name'] for cat in cats}
imageToAnno = defaultdict(list)
for anno in annotations:
    imageToAnno[anno['image_id']].append(anno)
data = []
for image in images:
    imageID = image['id']
    image_path = image['file_name']
    imageAnnos = imageToAnno[imageID]
    for anno in imageAnnos:
        catID = anno['category_id']
        name = catMap[catID]
        box = anno['bbox']
        seg = anno.get('segmentation', None)
        crowd = anno.get('iscrowd', 0)
        data.append({
            "Image Path": image_path,
            "Category ID": catID,
            "Category Name": name,
            "Bounding Box": box,
            "Segmentation": seg,
            "Is Crowd": crowd
        })
df = pd.DataFrame(data)
df
```
